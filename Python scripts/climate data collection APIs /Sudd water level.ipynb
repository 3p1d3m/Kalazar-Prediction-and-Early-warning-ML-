```python
import requests
import json
import csv
import os

# API URL
url = "https://dahiti.dgfi.tum.de/api/v2/download-water-level/"

# Securely use API key
api_key = "YOUR_API_KEY_HERE"  # Replace with your actual API key

# Request parameters
params = {
    "api_key": '23B90E3726570CCE2618C5DA159BE3F1F2891BE1CE6655F9EA7677923DD80327',
    "dahiti_id": 17749,  # White Nile station ID
    "format": "json"
}

# Folder to save CSV file
save_folder = "(SUDD)_water_level_data"
os.makedirs(save_folder, exist_ok=True)  # Create folder if it doesn't exist

# Send POST request
response = requests.post(url, json=params)

# Check response
if response.status_code == 200:
    data = response.json()
    
    # 🔍 Debugging: Print response structure
    print(json.dumps(data, indent=4))
    
    # Extract water level data
    if "data" in data:
        csv_file = os.path.join(save_folder, "white_nile_water_level.csv")
        
        # Save data to CSV
        with open(csv_file, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["Date", "Water Level", "Error", "Source"])  # CSV Header
            
            for entry in data["data"]:
                # 🔹 Check available keys and extract data safely
                date = entry.get("date") or entry.get("datetime") or "N/A"
                height = entry.get("height") or entry.get("value") or entry.get("water_level") or "N/A"
                error = entry.get("error", "N/A")
                source = entry.get("data", "N/A")
                
                writer.writerow([date, height, error, source])
        
        print(f"✅ Data saved successfully to {csv_file}")
    else:
        print("⚠️ No water level data found in response.")
    
elif response.status_code == 403:
    print("❌ Access denied! Check if your API key is correct and authorized.")
elif response.status_code == 400:
    print("❌ Bad request! Check if the parameter names are correct.")
else:
    print(f"❌ Error {response.status_code}: {response.text}")
```
